```python
import pandas as pd
import numpy as np

data = {
    'Order ID': ['CA-001','CA-002','CA-003','CA-004','CA-005'],
    'Category': ['Furniture','Technology','Office Supplies','Furniture','Technology'],
    'Sub-Category': ['Chairs','Phones','Binders','Tables','Laptops'],
    'Region': ['East','West','Central','South','East'],
    'Sales': [250.5, 890.0, 45.3, 320.8, 1200.0],
    'Profit': [41.9, 267.0, 8.1, -87.0, 360.0],
    'Quantity': [2, 3, 5, 1, 4]
}

df = pd.DataFrame(data)
df.head()
```

```python
df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
```

```python
df.groupby('Region')['Profit'].sum().sort_values(ascending=False)
```

```python
df.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False)
```

```python
print('Total Sales:', df['Sales'].sum())
print('Total Profit:', df['Profit'].sum())
print('Average Order Value:', df['Sales'].mean())
```

```python
import matplotlib.pyplot as plt

df.groupby('Category')['Sales'].sum().plot(kind='bar', color='steelblue')
plt.title('Sales by Category')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df.groupby('Region')['Profit'].sum().plot(kind='bar', color='green')
plt.title('Profit by Region')
plt.xlabel('Region')
plt.ylabel('Profit')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.scatter(df['Sales'], df['Profit'], color='coral')
plt.title('Sales vs Profit')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.tight_layout()
plt.show()
```
